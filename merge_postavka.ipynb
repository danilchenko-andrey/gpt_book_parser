# Match tilda to shopify format

```python
class FieldTransformer(object):
    def __init__(self, source_fields=[], default_value="", after=None):
        self.source_fields = source_fields
        self.default_value = default_value
        self.after = after if after else lambda x: x

    def transform(self, row):
        try:
            for source_field in self.source_fields:
                if source_field in row:
                    value = row[source_field]
                    if value and str(value).lower() not in ["nan", "none", "null", "not_found", "нет"]:
                        return self.after(value)
            return self.after(self.default_value)
        except Exception as e:
            print(f"Error in transform: row {row}, {e}")
            raise e

    @staticmethod
    def none():
        return FieldTransformer()
    
    @staticmethod
    def const(const_value):
        return FieldTransformer(default_value=const_value)
    
    @staticmethod
    def priority(source_fields, default_value="", after=None):
        return FieldTransformer(source_fields, default_value, after=after)
    
    @staticmethod
    def concat(source_fields, concat=" - ", default_value=""):
        return FieldTransformer(source_fields, concat, default_value)
```
